## Conditional Filtering

- Typically in data analysis our datasets are large enough that we don't filter based on position, but instead based on a condition.
- Conditional filtering allows us to select rows based on a condition on a column.
- This leads to a discussion on organizing our data.

- Conditional Filtering:
    - Filter by single condition
    - Filter by multiple condition
    - Check against multiple possible values

```python
df = pd.read_csv('Files/tips.csv')
```

```python
df.head()
```

```python
df[df['total_bill'] >= 40]
```

```python
df[(df['total_bill'] >= 40) & (df['sex'] == 'Male')]
```

```python
df[(df['total_bill'] >= 40) & (df['sex'] == 'Female')]
```

```python
df[(df['total_bill'] >= 30) | (df['sex'] == 'Male')]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Fri')]
```

```python
df[df['day'].isin(['Sun', 'Sat', 'Fri'])]
```

## Useful Methods

- While pandas has many built in methods, we can use the .apply() method call to apply any custom python function of our own to every row in a series.

### Part 1: Useful Methods - Apply on Single Column

```python
df = pd.read_csv('Files/tips.csv')
```

```python
df.head()
```

```python
def last_four(num):
    return str(num)[-4:]

df['CC Number'].apply(last_four)
```

```python
# Getting the last 4 digits of a credit card number stored in a column 'CC Number'
df['CC Number'].apply(lambda x: str(x)[-4:])
```

### Part 2: Useful Methods - Apply on Multiple Columns

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

### Missing Data

Options for missing data
- Keep it
- Remove it
- Replace it

### Keeping the missing data
Pros
- Easiest to do
- Does not manipulate or change the true data

Cons
- Many methods do not support NaN
- Often there are reasonable guesses 

### Dropping or Removing the missing data
Pros
- Easy to do
- Can be based on rules

Cons
- Potential to lose a lot of data or useful information
- Limits trained models for future data

### Filling in the missing data
Pros
- Potential to save a lot of data for use in training a model

Cons
- Hardest to do and somewhat arbitrary
- Potential to lead to false conclusions

### Group By Operations

A groupby() operation allows us to examine data on a per category basis. 

```python
pd.set_option('Display.max_rows', None)
```

```python
df = pd.read_csv('Files/mpg.csv')
```

```python
df.head()
```

```python
df.groupby(by = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'origin', 'name'])['model_year'].mean()
```

```python
df.groupby('origin')['name'].count()
```

Aggregation Functions which can be used on group by 

- mean(): Compute mean of groups
- sum(): Compute sum of group values
- size(): Compute group sizes
- count(): Compute count of groups
- std(): Standard deviation of groups
- var(): Compute variance of groups
- sem(): Standard error of the mean of groups
- describe(): Generates descriptive statistics
- first(): Compute first of group values
- last(): Compute last of group values
- nth: Take nth value or a subset if n is a list
- min(): Compute min of group values
- max(): Compute max of group values
