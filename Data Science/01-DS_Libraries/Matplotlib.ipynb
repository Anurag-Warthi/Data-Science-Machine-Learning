### Implementing Figures and Axes

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x
```

```python
y
```

```python
fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b);
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)


# Small Axes
axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])
axes2.plot(x, y);
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)


# Small Axes
axes2 = fig.add_axes([0.1, 0.2, 0.5, 0.5])
axes2.plot(x, y);
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)


# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x, y);
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([1, 1, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([1.1, 0, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
# dpi is used for clarity. higher the dpi value, more is the clarity
fig = plt.figure(figsize=(6, 6), dpi=200)

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a, b)
```
